```python
df = data.dropna(subset=['user_id_maped','duration_seconds','program_name'])
```

```python
summary = df.groupby(['program_name', 'program_class']).agg(
    total_views = ('user_id_maped', 'count'),
    unique_class=('user_id_maped', 'nunique'),
    total_watch_time = ('duration_seconds', 'sum')
).reset_index()

summary
```
